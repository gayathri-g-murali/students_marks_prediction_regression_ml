## FEATURE ENGINEERING - MODEL CREATION AND EVALUATION

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression , Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
df.head(2)
```

```python
features = ["study_hours_per_day","attendance_percentage","mental_health_rating","sleep_hours","part_time_job"]
```

```python
target = "exam_score"
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])	
```

```python
df_model
```

```python
# Splitting X and Y 
x = df_model[features]
```

```python
y = df_model[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
len(y_test)
```

```python
len(y_train)
```

```python
# Creating Models
models = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "DecisionTree": {
        "model": DecisionTreeRegressor(),
        "params": {
            "max_depth": [3, 5, 10],
            "min_samples_split": [2, 5]
        }
    },
    "RandomForest": {
        "model": RandomForestRegressor(),
        "params": {
            "n_estimators": [50, 100],
            "max_depth": [5, 10]
        }
    }
}
```

```python
# Best Models
best_models = []
```

```python
for name, config in models.items():
    print(f"Training {name}")
    grid = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")
    grid.fit(x_train, y_train)
    y_pred = grid.predict(x_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred)) 
    r2 = r2_score(y_test, y_pred)
    
    best_models.append({
        "model": name,
        "best_params": grid.best_params_,
        "rmse": rmse,
        "r2": r2
    })
```

```python
best_models
```

```python
result_df = pd.DataFrame(best_models)
```

```python
result_df.sort_values(by = "rmse")
```

```python
import joblib

best_row = result_df.sort_values(by ="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row["model"]
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model = best_model_config["model"]
```

```python
final_model.fit(x,y)
```

```python
# Exporting the model
joblib.dump(final_model,"best_model.pkl")
```
